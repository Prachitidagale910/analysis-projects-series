```python

axs = sns.countplot(x='gender', data = df, hue = 'gender')

for bars in axs.containers:
    axs.bar_label(bars)
```

```python
sales_gen = df.groupby(['gender'], as_index=False)['price'].sum().sort_values(by='price', ascending=False)
# Plotting the bar chart with hue
sns.barplot(x='gender', y='price', hue='gender', data=sales_gen, palette='Set2')

# Adding labels to each bar
for bars in axs.containers:
    axs.bar_label(bars)

# Show the plot
plt.show()
```

from the above we can conclude that female are the most buyers in the given malls...


```python


# Grouping data by category and summing the quantity
sales_gen = df.groupby(['category'], as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False)

# Plotting the bar chart with hue set to category
sns.barplot(x='category', y='quantity', hue='category', data=sales_gen, palette='Set2', legend=False)

# Adding labels to each bar
for bars in axs.containers:  # Change 'axs' to 'ax' if you're using a single Axes object
    axs.bar_label(bars)

# Adjusting the figure size
sns.set(rc={'figure.figsize': (2, 5)})

# Show the plot
plt.show()
```

This indicates a strong consumer preference for clothing, which could be useful for inventory and marketing strategies.

```python
plt.figure(figsize=(12, 6)) 

axs = sns.countplot(x='payment_method', data = df, hue = 'payment_method', palette='Set2')

for bars in axs.containers:
    axs.bar_label(bars)
```

```python
'''
Exploring more on invoice_date column
'''
```

```python
# Display unique values in the invoice_date column
print(df['invoice_date'].unique())
```

```python

# Check the data type of the invoice_date column
print(df['invoice_date'].dtype)
```

```python
# Display the first few rows of the invoice_date column
print(df['invoice_date'].head(10))
```

```python
# Ensure invoice_date is in datetime format using .loc
df.loc[:, 'invoice_date'] = pd.to_datetime(df['invoice_date'])

# Extract year and month using .loc
df.loc[:, 'year'] = df['invoice_date'].dt.year
df.loc[:, 'month'] = df['invoice_date'].dt.month

# Grouping by year and month to calculate total sales
monthly_sales = df.groupby(['year', 'month'], as_index=False)['price'].sum()

# Plotting the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='month', y='price', hue='year', marker='o')

# Adding titles and labels
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                        'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year')
plt.grid()
plt.show()
```

```python
# Convert invoice_date to datetime
df.loc[:, 'invoice_date'] = pd.to_datetime(df['invoice_date'])

# Extract year, month, day, and day of the week
df.loc[:, 'year'] = df['invoice_date'].dt.year
df.loc[:, 'month'] = df['invoice_date'].dt.month
df.loc[:, 'day'] = df['invoice_date'].dt.day
df.loc[:, 'day_of_week'] = df['invoice_date'].dt.day_name()
```

```python
monthly_sales = df.groupby(['year', 'month'], as_index=False)['price'].sum()
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='month', y='price', hue='year', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.show()
```

so in 2023 the sales are high and mostly in the june month.


```python
daily_sales = df.groupby('invoice_date')['price'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_sales, x='invoice_date', y='price')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python
peak_sales_dates = df.groupby('invoice_date')['price'].sum().sort_values(ascending=False).head(10)
print("Top 10 Sales Dates:\n", peak_sales_dates)
```

```python
# Set the figure size
plt.figure(figsize=(15, 6))  # Width: 15, Height: 6

# Create the histogram with a density plot
sns.histplot(df['invoice_date'], bins=30, kde=True)
plt.title('Sales Distribution Over Time')
plt.xlabel('Invoice Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-ticks for better visibility

# Show the plot
plt.show()
```
